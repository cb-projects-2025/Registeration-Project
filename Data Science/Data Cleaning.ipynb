```python
import pandas as pd

# Load the Excel file
excel_path = r"C:\Users\Charbel\Desktop\Data Science X5\Data Collection - Registration App Final2.xlsx"
excel_file = pd.ExcelFile(excel_path)

# Read all sheets into a dictionary of DataFrames
sheets_dict = {sheet_name: excel_file.parse(sheet_name) for sheet_name in excel_file.sheet_names}

# Display sheet names and the first few rows of each sheet
{sheet: df.head() for sheet, df in sheets_dict.items()}

payment_method_df = sheets_dict["Payment_Method"]
payment_df = sheets_dict["Payment"]
institution_df = sheets_dict["Institution"]
academic_level_df = sheets_dict["Academic_Level"]
track_df = sheets_dict["Track"]
batch_df = sheets_dict["Batch"]
track_batch_df = sheets_dict["Track_Batch"]
user_batch_df = sheets_dict["User_Batch"]
instructor_batch_df = sheets_dict["Instructor_Batch"]
user_df = sheets_dict["USER"]
```

```python
# Clean and preprocess each DataFrame: drop duplicates and handle missing values (simple strategy: drop for now)

cleaned_dfs = {}
for name, df in sheets_dict.items():
    cleaned_df = df.drop_duplicates().dropna(how='all')  
    cleaned_dfs[name] = cleaned_df

# Assign each cleaned DataFrame to a variable
payment_method_df = cleaned_dfs["Payment_Method"]
payment_df = cleaned_dfs["Payment"]
institution_df = cleaned_dfs["Institution"]
academic_level_df = cleaned_dfs["Academic_Level"]
track_df = cleaned_dfs["Track"]
batch_df = cleaned_dfs["Batch"]
track_batch_df = cleaned_dfs["Track_Batch"]
user_batch_df = cleaned_dfs["User_Batch"]
instructor_batch_df = cleaned_dfs["Instructor_Batch"]
user_df = cleaned_dfs["USER"]

# Show the number of rows removed for each DataFrame
rows_removed = {
    name: len(sheets_dict[name]) - len(cleaned_dfs[name])
    for name in sheets_dict
}

rows_removed
```

```python
# Check for null values in each DataFrame
null_counts = {
    name: df.isnull().sum()
    for name, df in cleaned_dfs.items()
}

# Filter to show only columns with nulls
null_columns_only = {
    name: counts[counts > 0]
    for name, counts in null_counts.items()
    if counts.sum() > 0
}

null_columns_only
```
